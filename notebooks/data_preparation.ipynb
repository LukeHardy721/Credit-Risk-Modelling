# Data Preparation

### Importing and Filtering

```python
import pandas as pd
import numpy as np

file_path = "../data/raw/accepted_2007_to_2018Q4.csv"
df = pd.read_csv(file_path)
```

```python
pd.set_option('display.max_columns', None)
```

```python
print(df.shape)
df.head()
```

```python
df['loan_status'].value_counts()
```

Let's filter for the values of loan_status we are interested in.

```python
df = df[df['loan_status'].isin(["Fully Paid", "Charged Off"])]
df['target'] = df['loan_status'].map({'Fully Paid': 0, 'Charged Off': 1})
print(df.shape)
```

### Stripping Whitespace and Dropping Columns

```python
df.columns = df.columns.str.strip()
```

```python
# Drop columns with >80% missing values
df.drop(columns=df.columns[df.isnull().mean() > 0.8], inplace=True)

# Drop columns with only one unique value
df.drop(columns=df.columns[df.nunique() <= 1], inplace=True)
```

```python
print(df.shape)
df.head()
```

```python
# We need to drop any columns that contain data from after the loan was approved, as we can only predict using data we have available at the time of making the prediction.
leakage_cols = [
    'funded_amnt','funded_amnt_inv','issue_d','total_pymnt', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 'recoveries',
    'collection_recovery_fee', 'last_pymnt_d', 'last_pymnt_amnt', 'last_credit_pull_d', 'last_fico_range_high', 'last_fico_range_low', 'next_pymnt_d', 'out_prncp', 'out_prncp_inv',
    'disbursement_method', 'hardship_flag', 'settlement_status', 'debt_settlement_flag'
]

# We can also safely drop columns that are clearly redundant, such as 'id', 'url' and 'loan_status' (since we have converted it to 'target').
redundant_cols = ['id','url','loan_status']

cols_to_drop = leakage_cols + redundant_cols
df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)
df.shape
```

### Converting Data Types and Fixing Formatting

```python
df.dtypes.value_counts()
```

```python
# Numerical features
numerical_cols = df.select_dtypes(include=['number', 'bool']).columns.tolist()
print(numerical_cols)
print(len(numerical_cols))

df[numerical_cols].head()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print(categorical_cols)
print(len(categorical_cols))

df[categorical_cols].head()
```

```python
datetime_cols = df.select_dtypes(include=['datetime64[ns]']).columns.tolist()
print(datetime_cols)
```

```python
print(df['term'].value_counts())
print()
print(df['emp_length'].value_counts())
```

```python
import re

df['term'] = df['term'].str.extract(r'(\d+)').astype(float)

def parse_emp_length(val):
    if pd.isnull(val):
        return np.nan
    elif '< 1' in val:
        return 0
    elif '10+' in val:
        return 10
    else:
        match = re.search(r'(\d+)', val)
        return float(match.group(1)) if match else np.nan

df['emp_length'] = df['emp_length'].apply(parse_emp_length)
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%Y', errors='coerce') # Force errors to be NaT if not a valid datetime
```
